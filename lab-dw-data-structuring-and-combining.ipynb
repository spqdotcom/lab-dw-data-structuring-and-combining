```python
def date_numbers(x):
    # Convertir a string si no lo es
    if not isinstance(x, str):
        x = str(x)

    if x == '0':
        return x

    try:
        number = x.split('/')[1]  # Suponiendo que siempre esperas una cadena con una barra `/`
        return int(number)
    except (ValueError, TypeError, IndexError):
        # Manejo de posibles errores que podrían surgir
        return None
```

```python
df["number_of_open_complaints_bis"] = df["number_of_open_complaints"].apply(date_numbers)
```
